```python
# Dividir em ficheiros mais pequenos, neste caso, com 300000 linhas
chunk_size=300000
batch_no=1
for chunk in pd.read_csv("global_climate_data/ghcnd_daily.csv",chunksize=chunk_size,iterator=True,low_memory=False):
    chunk.to_sql('ghcnd',conn, if_exists='append')
    batch_no+=1
    print('index: {}'.format(batch_no))
```
