```python
#Import library´s
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from geopy import distance
import plotly.graph_objects as go
import folium
```

```python
# create a DataFrame with city names, latitudes, and longitudes
df = pd.DataFrame({'name': ['viseu', 'coimbra', 'leiria', 'tomar', 'santarém','aveiro','soure','odemira', 'lagoa', 'faro']})
```

```python
geolocator = Nominatim(user_agent="geo_IPT")

geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df['location'] = df['name'].apply(geocode)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)

print(df['name'])
```

```python
df1 = pd.DataFrame({
    'City': df['name'],
    'Latitude': df['point'].str[0],
    'Longitude': df['point'].str[1]
})
# create a map object
m = folium.Map(location=[39.3, -8], zoom_start=7)

# add a marker for each city to the map
for i, row in df1.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=row['City']).add_to(m)

folium.Marker(
    location=[37.5215, -8.6261],
    popup="Referência: city",
    icon=folium.Icon(color="green")
).add_to(m)

# display the map
m
```
