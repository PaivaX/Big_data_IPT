### Determine a percentagem de null values em cada uma das variáveis.

### Crie uma nova coluna denominada count_nan que registe para cada observação o número de dias com temperaturas máximas nulas (NaN).

```python
for i, row in ghcnd.iterrows():
    days_nan = 0
    for x in range(1, 32): # 32 is not included
        if pd.isna(ghcnd.loc[i, f'value{x}']):
            days_nan += 1
    ghcnd.loc[i, 'count_nan'] = days_nan

ghcnd
```

### Crie um novo dataframe que considere apenas os registos com um valor total de NaN inferior a 7.

### Reccorra agora a chunking para determinar a percentagem de null values em cada uma das variáveis na totalidade dos registos do dataframe.

```python
run = False # change to True to run

if run:
    # Dividir em ficheiros mais pequenos, neste caso, com 300000 linhas
    chunk_size = 300000

    global_null_counts = None
    row_count = 0
    batch = 1
    for df_chunk in pd.read_csv(ghcnd_daily_csv_path, chunksize=chunk_size, low_memory=False):
        # calculate the null values for each column
        null_counts = pd.DataFrame(df_chunk.isnull().sum())
        if global_null_counts is None:
            global_null_counts = null_counts
        else:
            global_null_counts += null_counts

        row_count += len(df_chunk)
        
        print(f'processed batch # {batch}')
        batch += 1
    
    # divide values by total values to get percentages
    global_null_counts[0] = global_null_counts[0] / row_count

global_null_counts
```

### Determine a temperatura média para cada observação (estação/ano/mês) do dataframe. Guarde os valores obtidos numa nova coluna denominada daily_avg_temp. Não se esqueça de considerar apenas as colunas cujo nome começa por value.

```python
run = False # change to True to run

if run:
    def mean_without_neg9999(row):
        filtered_row = row[row != -9999]  # filter out -9999 values
        if len(filtered_row) == 0:  # handle case where all values are -9999
            return np.nan
        else:
            return filtered_row.mean()
        
    daily_avg_tmp_csv_path = os.path.join(ignored_dir, 'daily_avg_temp.csv')
    if os.path.exists(daily_avg_tmp_csv_path):
        os.remove(daily_avg_tmp_csv_path) # clear csv
        
    # Dividir em ficheiros mais pequenos, neste caso, com 300000 linhas
    chunk_size = 300000

    batch = 1
    for df_chunk in pd.read_csv(ghcnd_daily_csv_path, chunksize=chunk_size, low_memory=False):
        df_chunk['daily_avg_temp'] = df_chunk.apply(lambda row: mean_without_neg9999(row.filter(like='value')), axis=1)

        df_chunk.to_csv(daily_avg_tmp_csv_path, mode='a', header=(True if batch == 1 else False), index=False)
        
        print(f'processed batch # {batch}')
        batch += 1

df_chunk
```
