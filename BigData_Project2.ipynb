### Determine a percentagem de null values em cada uma das variáveis.

```python
#
conn = sql.connect(patch + "ghcnd_daily.db")
print('Conexão com a BD com sucesso.')

#
df = pd.read_sql_query('SELECT id, year, month, element, value1, value2, value3, value4, value5, value6, value7, value8, value9, value10,value11, value12, value13, value14, value15, value16, value17, value18, value19, value20, value21, value22, value23, value24, value25, value26, value27, value28, value29, value30, value31 FROM ghcnd where (id="Horta" OR id="Funchal" OR id="Lisboa" OR id="Castelo Branco" OR id="Faro")',conn)


# Determine a porcentagem de valores nulos em cada coluna que são iguais a -9999
for col in df.columns:
    pct_null = df.loc[df[col] == -9999, col].shape[0] / df.shape[0] * 100
    print(f"A percentagem de valores nulos na coluna {col} é de {pct_null:.2f}%.")
    
conn.close()
```

### Crie uma nova coluna denominada count_nan que registe para cada observação o número de dias com temperaturas máximas nulas (NaN).

```python
# Conecte-se ao banco de dados e carregue os dados em um dataframe
conn = sql.connect(patch + "ghcnd_daily.db")
print('Conexão com a BD com sucesso.')

#
cursor = conn.cursor()

#
df = pd.read_sql_query('SELECT id, year, month, element, value1, value2, value3, value4, value5, value6, value7, value8, value9, value10,value11, value12, value13, value14, value15, value16, value17, value18, value19, value20, value21, value22, value23, value24, value25, value26, value27, value28, value29, value30, value31 FROM ghcnd where (id="Horta" OR id="Funchal" OR id="Lisboa" OR id="Castelo Branco" OR id="Faro")',conn)

#
df["count_nan"] = df.loc[:, "value1":"value31"].isin([-9999]).sum(axis=1)

# Adicionar a nova coluna ao banco de dados
df.to_sql("ghcnd", conn, if_exists="replace", index=False)

# Confirmar que a nova coluna foi adicionada ao banco de dados
cursor.execute("PRAGMA table_info(ghcnd)")
print(cursor.fetchall())

# Fechar cursor e a conexão com a BD
cursor.close()

# Exiba as primeiras linhas do dataframe com a nova coluna "count_nan"
print(df.head())
```

### Crie um novo dataframe que considere apenas os registos com um valor total de NaN inferior a 7.

```python
# Conecte-se ao banco de dados e carregue os dados em um dataframe
conn = sql.connect(patch + "ghcnd_daily.db")
print('Conexão com a BD com sucesso.')


# Criar um dataframe booleano indicando os valores NaN
df_nan = df.loc[:, "value1":"value31"].isna()

# Somar o número de NaN para cada linha
num_na = df_nan.sum(axis=1)

# Criar um novo dataframe que contém apenas as linhas com um número de NaN inferior a 7
df1 = df[num_na < 7]

# Exiba as primeiras linhas do dataframe
print(df1.head())
```

### Agrupe os dados por nome de estação e ano aplicando a função mean. Mostre apenas os resultados para a coluna daily_avg_temp.
